# Organize

```python
X=np.linspace(0,10,100)
Y1, Y2=np.sin(X), np.cos(X)
plt.plot(X, Y1, X, Y2)
```

```python
fig, (ax1, ax2) = plt.subplots(2,1)
ax1.plot(X, Y1, color="C1")
ax2.plot(X, Y2, color="C0")
```

```python
fig, (ax1, ax2) = plt.subplots(1,2)
ax1.plot(Y1, X, color="C1")
ax2.plot(Y2, X, color="C0")
```

# Label
